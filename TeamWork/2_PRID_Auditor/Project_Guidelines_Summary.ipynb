# MP607: Major Project Execution Roadmap

## 1. Project Phase 1: Initiation and Planning

### Q1: What are the primary objectives of the Major Project phase?
**Answer:**
- **Skill Application:** To apply theoretical knowledge to solve real-world engineering problems.
- **Innovation:** To research and implement creative solutions in modern domains (AI, IoT, Cloud, Security).
- **Collaboration:** To work effectively in a team (max 5 members) and follow industrial development standards.
- **Documentation:** To learn professional report writing and project presentation techniques.

### Q2: List the critical milestones for project approval.
**Answer:**
1. **Domain Selection:** Identifying a problem area (e.g., Cybersecurity, Web App).
2. **Guide Allocation:** Finalizing a faculty supervisor to oversee technical progress.
3. **Synopsis Submission:** A 2-3 page document defining objectives, methodology, and tools.
4. **Approval Presentation:** Defending the project's feasibility and scope before a department panel.

## 2. Project Phase 2: Design and Development

### Q3: What diagrams must be included in the project design document?
**Answer:**
- **System Architecture:** A high-level overview of components (e.g., Frontend, Backend, Database).
- **UML Diagrams:** 
  - **Use Case Diagram:** Shows interactions between actors and the system.
  - **Class Diagram:** Defines static structure and relationships.
  - **Sequence Diagram:** Shows object interactions over time.
- **ER Diagram:** Database schema design.
- **Flowcharts:** Logic for critical backend algorithms.

### Q4: Explain the importance of version control in a group project.
**Answer:**
- **Conflict Prevention:** Tools like Git allow multiple students to work on different parts of the code simultaneously (e.g., the PRID framework we used).
- **Recovery:** Ability to revert to previous working states if a new feature breaks the system.
- **Audit Trail:** Clear visibility of who contributed what, which helps in the final Viva assessment.

## 3. Project Phase 3: Testing, Deployment, and Final Viva

### Q5: What are the key components of the final project report?
**Answer:**
1. **Introduction:** Background and Problem Statement.
2. **Literature Survey:** Review of existing solutions and technologies.
3. **Requirements Analysis:** SRS (Software Requirement Specification).
4. **Design:** Architectural and UML diagrams.
5. **Implementation:** Code snippets, screenshots of the UI.
6. **Testing:** Test cases (Unit, Integration, System) and results.
7. **Conclusion & Future Scope:** What was achieved and what can be improved.

### Q6: How should a student prepare for the External Viva Voce?
**Answer:**
- **Individual Contribution:** Be ready to explain exactly what *you* built (your PRID zone).
- **Code Walkthrough:** Understand every line of the core logic, even if a teammate wrote it.
- **Theoretical Alignment:** Relate project features to course subjects (e.g., "We used RSA for security as studied in DIS").
- **Demonstration:** Ensure the project is locally hosted and ready for a live demo during the exam.
